```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.goodreads.com/quotes/tag/inspirational'

# Define headers to mimic a real browser request
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36',
}


response = requests.get(url, headers=headers)

# Check if the request was successful
if response.status_code == 200:
    # Parse the HTML content using BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all quote elements on the page
    quotes = soup.find_all('div', class_='quoteDetails')  # This class contains quote, author, and genre

    # Loop through the quotes and extract the desired information
    for quote in quotes:
        # Extract the quote text
        quote_text = quote.find('div', class_='quoteText').text.strip()

        # Extract the author of the quote
        author = quote.find('span', class_='authorOrTitle').text.strip()

        # Extract the genres (tags)
        genres = [tag.text.strip() for tag in quote.find_all('a', class_='smallText')]

        # Print the extracted data
        print(f"Quote: {quote_text}")
        print(f"Author: {author}")
        print(f"Genres: {', '.join(genres)}")
        print('-' * 60)
      
else:
    print(f"Failed to retrieve the page. Status code: {response.status_code}")
```
